```python
# Dependencies and Setup
import pandas as pd
import os

# File to Load (Remember to Change These)
school_data_path = os.path.join("Resources", "schoolData.csv")
student_data_path = os.path.join("Resources", "studentData.csv")

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_path)
student_data = pd.read_csv(student_data_path)

# Combine the data into a single dataset.  
fd = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])

fd.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#calc number of unique schools
num_schools = fd.school_name.nunique()
#print(num_schools)

#calc number of students (just number of rows)
num_students = len(fd.student_name)
#print(num_students)

#calc total budget
tot_budget = fd.budget.unique().sum()
print(tot_budget)

#calc avg math score
avg_math = fd.math_score.mean()
#print(avg_math)

#calc avg reading score
avg_reading = fd.reading_score.mean()
#print(avg_reading)

#filter to find students passing math
passing_math = fd[fd.math_score >= 70]
#calculate percent passing math
pct_passing_math = (len(passing_math.student_name) / num_students)*100
#print(pct_passing_math)

#filter to find students passing reading
passing_reading = fd[fd.reading_score >= 70]
#calculate percent passing reading
pct_passing_reading = (len(passing_reading.student_name) / num_students)*100
#print(pct_passing_reading)

#filter the data frame of studetns passing math further to those also passing reading
passing_ovr = passing_math[passing_math.reading_score >= 70]
pct_passing_ovr = (len(passing_ovr.student_name) / num_students) * 100
#print(pct_passing_ovr)

#####################################################################################
#create data fram with calculated fields
district_summary = pd.DataFrame({
    "Total Schools": [num_schools],
    "Total Students": [num_students],
    "Total Budget": ["$"+ f"{tot_budget:,d}"],
    "Average Math Score": [avg_math],
    "Average Reading Score": [avg_reading],
    "% Passing Math": [pct_passing_math],
    "% Passing Reading": [pct_passing_reading],
    "% Passing Overall": [pct_passing_ovr]
})
district_summary.head()
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
constants = fd[["school_name", "type", "budget"]].drop_duplicates()
constants = constants.rename({
    "type": "School Type",
    "budget": "Total School Budget"
})
#print(constants.shape)

grouped = fd.groupby(["school_name"]).agg({
    "student_name": "count",
    "math_score": "mean",
    "reading_score": "mean"    
})
grouped = grouped.rename(columns = {
    "student_name": "Total Students",
    "math_score": "Average Math Score",
    "reading_score" : "Average Reading Score"
})
#print(grouped.head())

passedMath = passing_math.groupby("school_name").agg({
    "student_name": "count"
})
passedMath = passedMath.rename(columns = {"student_name": "#Math"})
#print(passedMath.head())

passedReading = passing_reading.groupby("school_name").agg({
    "student_name": "count"
})
passedReading = passedReading.rename(columns = {"student_name": "#Reading"})
#print(passedReading.head())

passedOverall = passing_ovr.groupby("school_name").agg({
    "student_name": "count"
})
passedOverall = passedOverall.rename(columns = {"student_name": "#Overall"})
#print(passedOverall.head())

fullData = pd.merge(grouped, passedMath, how="left", on=["school_name"])
fullData = pd.merge(fullData, passedReading, how="left", on=["school_name"])
fullData = pd.merge(fullData, passedOverall, how="left", on=["school_name"])
fullData = pd.merge(fullData, constants, how="left", on=["school_name"])
fullData["Per Student Budget"] = (fullData["budget"] / fullData["Total Students"])
fullData["% Passing Math"] = (fullData["#Math"] / fullData["Total Students"]) *100
fullData["% Passing Reading"] = (fullData["#Reading"] / fullData["Total Students"]) *100
fullData["% Overall Passing"] = (fullData["#Overall"] / fullData["Total Students"]) *100
#fullData.set_index("school_name")
#fullData.head()

finalData = fullData[["school_name","type","Total Students", "budget","Per Student Budget", "Average Math Score",
                      "Average Reading Score", "% Passing Math", "% Passing Reading", "% Overall Passing"]]
finalData = finalData.set_index("school_name")

finalData = finalData.rename(columns = {
    "type":"School Type",
    "budget": "Total School Budget"
})
finalData
```

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
grade9 = fd[fd.grade == "9th"]
grade9Grouped = grade9.groupby("school_name").agg({"math_score": "mean"})
#grade9Grouped.head()
grade10 = fd[fd.grade == "10th"]
grade10Grouped = grade10.groupby("school_name").agg({"math_score": "mean"})
#grade10Grouped.head()
grade11 = fd[fd.grade == "11th"]
grade11Grouped = grade11.groupby("school_name").agg({"math_score": "mean"})
#grade11Grouped.head()
grade12 = fd[fd.grade == "12th"]
grade12Grouped = grade12.groupby("school_name").agg({"math_score": "mean"})
grade12Grouped.head()

mathByGrade = pd.merge(grade9Grouped, grade10Grouped, how = "left", on = "school_name")
mathByGrade = mathByGrade.rename(columns = {
    "math_score_x": "9th",
    "math_score_y": "10th"
})
mathByGrade = pd.merge(mathByGrade, grade11Grouped, how = "left", on = "school_name")
mathByGrade = pd.merge(mathByGrade, grade12Grouped, how = "left", on = "school_name")
mathByGrade = mathByGrade.rename(columns = {
    "math_score_x": "11th",
    "math_score_y": "12th"
})

mathByGrade
```

## Reading Score by Grade 

* Perform the same operations as above for reading scores

```python
grade9r = fd[fd.grade == "9th"]
grade9rGrouped = grade9r.groupby("school_name").agg({"reading_score": "mean"})
#grade9Grouped.head()
grade10r = fd[fd.grade == "10th"]
grade10rGrouped = grade10r.groupby("school_name").agg({"reading_score": "mean"})
#grade10Grouped.head()
grade11r = fd[fd.grade == "11th"]
grade11rGrouped = grade11r.groupby("school_name").agg({"reading_score": "mean"})
#grade11Grouped.head()
grade12r = fd[fd.grade == "12th"]
grade12rGrouped = grade12r.groupby("school_name").agg({"reading_score": "mean"})
grade12rGrouped.head()

readingByGrade = pd.merge(grade9rGrouped, grade10rGrouped, how = "left", on = "school_name")
readingByGrade = readingByGrade.rename(columns = {
    "reading_score_x": "9th",
    "reading_score_y": "10th"
})
readingByGrade = pd.merge(readingByGrade, grade11rGrouped, how = "left", on = "school_name")
readingByGrade = pd.merge(readingByGrade, grade12rGrouped, how = "left", on = "school_name")
readingByGrade = readingByGrade.rename(columns = {
    "reading_score_x": "11th",
    "reading_score_y": "12th"
})

readingByGrade
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
bins = [0, 585, 630, 645, 675]
bin_labels = ["<$585", "$585-629", "$630-644", "$645-675"]
perStudent = finalData[["Per Student Budget"]]
withPerStudent = pd.merge(fd, perStudent, how = "left", on = "school_name")
withPerStudent["budget_bin"] = pd.cut(withPerStudent["Per Student Budget"], bins, labels=bin_labels)
binGrouped = withPerStudent.groupby("budget_bin").agg({
    "student_name": "count",
    "math_score": "mean",
    "reading_score": "mean"
})
binGrouped = binGrouped.rename(columns = {
    "student_name": "tot_students"
})
#print(binGrouped)
binPassingMath = withPerStudent[withPerStudent.math_score >= 70].groupby("budget_bin").agg({"student_name": "count"})
binPassingMath = binPassingMath.rename(columns = {"student_name": "num_passing_math"})
#print(binPassingMath)
binPassingReading = withPerStudent[withPerStudent.reading_score >= 70].groupby("budget_bin").agg({"student_name": "count"})
binPassingReading = binPassingReading.rename(columns = {"student_name": "num_passing_reading"})
#print(binPassingReading)
binPassingOvr = withPerStudent[(withPerStudent.math_score >= 70) & (withPerStudent.reading_score >= 70)].groupby("budget_bin").agg({"student_name": "count"})
binPassingOvr = binPassingOvr.rename(columns = {"student_name": "num_passing_ovr"})
#print(binPassingOvr)

schoolSpending = pd.merge(binGrouped, binPassingMath, how = "left", on = "budget_bin")
schoolSpending = pd.merge(schoolSpending, binPassingReading, how = "left", on = "budget_bin")
schoolSpending = pd.merge(schoolSpending, binPassingOvr, how = "left", on = "budget_bin")
schoolSpending["% Passing Math"] = (schoolSpending.num_passing_math / schoolSpending.tot_students)*100
schoolSpending["% Passing Reading"] = (schoolSpending.num_passing_reading / schoolSpending.tot_students)*100
schoolSpending["% Overall Passing"] = (schoolSpending.num_passing_ovr / schoolSpending.tot_students)*100
schoolSpendingFiltered = schoolSpending[[
    "math_score", "reading_score", "% Passing Math", "% Passing Reading", "% Overall Passing" 
]]

schoolSpendingFiltered = schoolSpendingFiltered.rename(index = {
    "budget_bin": "Spending Ranges (Per Student)",
},
                                                       columns = {
    "math_score": "Average Math Score",
    "reading_score" : "Average Reading Score"
})
schoolSpendingFiltered
```

## Scores by School Size

* Perform the same operations as above, based on school size.

```python
size_bins = [0, 1000, 2000, 5000]
size_bin_labels = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
sizeStart = fd[["school_name", "student_name", "size", "reading_score", "math_score"]]
sizeStart["School Size"] = pd.cut(sizeStart["size"],size_bins, labels = size_bin_labels )
sizeBinned = sizeStart.groupby("School Size").agg({
    "student_name": "count",
    "math_score": "mean",
    "reading_score": "mean"
})
sizeBinned = sizeBinned.rename(columns = {"student_name": "tot_students"})
sizePassingMath = sizeStart[sizeStart.math_score >= 70].groupby("School Size").agg({"student_name": "count"})
sizePassingMath = sizePassingMath.rename(columns = {"student_name": "num_passing_math"})
#sizePassingMath.head()
sizePassingReading = sizeStart[sizeStart.reading_score >= 70].groupby("School Size").agg({"student_name": "count"})
sizePassingReading = sizePassingReading.rename(columns = {"student_name": "num_passing_reading"})
sizePassingOvr = sizeStart[(sizeStart.math_score >= 70) & (sizeStart.reading_score >= 70)].groupby("School Size").agg({"student_name": "count"})
sizePassingOvr = sizePassingOvr.rename(columns = {"student_name": "num_passing_ovr"})

schoolSize = pd.merge(sizeBinned, sizePassingMath, how = "left", on = "School Size")
schoolSize = pd.merge(schoolSize, sizePassingReading, how = "left", on = "School Size")
schoolSize = pd.merge(schoolSize, sizePassingOvr, how = "left", on = "School Size")

schoolSize["% Passing Math"] = (schoolSize.num_passing_math / schoolSize.tot_students) * 100
schoolSize["% Passing Reading"] = (schoolSize.num_passing_reading / schoolSize.tot_students) * 100
schoolSize["% Overall Passing"] = (schoolSize.num_passing_ovr / schoolSize.tot_students) * 100
schoolSizeFiltered = schoolSize[[
    "math_score", "reading_score", "% Passing Math", "% Passing Reading", "% Overall Passing"
]]
schoolSizeFiltered = schoolSizeFiltered.rename(columns = {
    "math_score": "Average Math Score",
    "reading_score": "Average Reading Score"
})
schoolSizeFiltered
```
